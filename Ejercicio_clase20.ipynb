**Identificación OUTLIERS**


1.   failures	(Número de asignaturas reprobadas previamente (0–4))
no son errores; simplemente indica que la mayoría no ha reprobado asignaturas.

2.   famrel	(Calidad de las relaciones familiares (1 - muy mala a 5 - excelente)
los valores reales estan dentro del rango

3.   Dalc	(Consumo de alcohol entre semana (1 - muy bajo a 5 - muy alto))
 Al igual que famrel, los valores reales están dentro de 1–5.

4.   absences	(Número de faltas a clases (0–93))
Esta variable es la más sensible; los estudiantes con más de 20 faltas son outliers reales que afectan al análisis si no se manejan.


```python
# Seleccionar variables de interés
vars_outliers = ["failures", "famrel", "Dalc", "absences"]

# Crear diccionario para guardar outliers
outliers_dict = {}

for col in vars_outliers:
    Q1 = df_students[col].quantile(0.25)
    Q3 = df_students[col].quantile(0.75)
    IQR = Q3 - Q1

    # Definir límites inferior y superior
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filas que tienen outliers
    outlier_rows = df_students[(df_students[col] < lower_bound) | (df_students[col] > upper_bound)]

    # Guardar resultados
    outliers_dict[col] = outlier_rows[[col]]

    print(f"{col} → Límites: {lower_bound:.2f} a {upper_bound:.2f} | Outliers: {len(outlier_rows)} filas")

# Opcional: ver todos los outliers de una variable
# outliers_dict["absences"]
```

##Tratamiento de datos

```python
#Trabajar outlier
# Calcular el percentil 95 de absences
upper_limit = df_students["absences"].quantile(0.95)
print(f"Percentil 95 de absences: {upper_limit}")

# Crear una nueva columna truncada en df_students
df_students["absences_trunc"] = df_students["absences"].apply(lambda x: upper_limit if x > upper_limit else x)

# Revisar los valores máximos antes y después
print("Máximo original:", df_students["absences"].max())
print("Máximo truncado:", df_students["absences_trunc"].max())
```

##División del conjunto de datos

```python
X= df_students.drop('G3', axis=1)
y= df_students['G3']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```

```python
#Normalización
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

##Modelos Supervisados

```python
#Regresión lineal
from sklearn.linear_model import LinearRegression

model_rl = LinearRegression() #Instancia modelo
model_rl.fit(X_train, y_train) #Entrenar el modelo
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = model_rl.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Error Absoluto Medio (MAE): {mae:.2f}")
print(f"Error Cuadrático Medio (MSE): {mse:.2f}")
print(f"Raíz del Error Cuadrático Medio (RMSE): {rmse:.2f}")
print(f"Coeficiente de Determinación (R²): {r2:.2f}")
```

**Resumen general**

Error promedio: ~1–2 puntos de diferencia → bastante preciso para predicciones
de notas.

R² alto (0.85) → modelo robusto y confiable para estimar la nota final.

```python
##KNN regresión
from sklearn.neighbors import KNeighborsRegressor

model_knn = KNeighborsRegressor(n_neighbors=5) #Instancia modelo
model_knn.fit(X_train, y_train) #Entrenar el modelo
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = model_knn.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Error Absoluto Medio (MAE): {mae:.2f}")
print(f"Error Cuadrático Medio (MSE): {mse:.2f}")
print(f"Raíz del Error Cuadrático Medio (RMSE): {rmse:.2f}")
print(f"Coeficiente de Determinación (R²): {r2:.2f}")
```

**Resumen**
Un RMSE ~2.89 indica que las predicciones pueden estar hasta 3 puntos fuera de la nota real.

Modelo explica el 64% de la variabilidad de las notas.

```python
#Random forest
from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor(n_estimators=100, random_state=42) #Instancia modelo
model_rf.fit(X_train, y_train) #Entrenar el modelo
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = model_rf.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Error Absoluto Medio (MAE): {mae:.2f}")
print(f"Error Cuadrático Medio (MSE): {mse:.2f}")
print(f"Raíz del Error Cuadrático Medio (RMSE): {rmse:.2f}")
print(f"Coeficiente de Determinación (R²): {r2:.2f}")
```

**Resumen**
Random Forest está supera ampliamente a los modelos anteriores (regresión lineal y KNN).

Predicciones muy cercanas a las notas reales y R² muy alto.

Además, puedes revisar la importancia de cada variable para ver cuáles influyen más en la nota final (G1, G2, failures, etc.)

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Aplicar Kmeans con 4 clusters
sse=[]
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

# Aplicar PCA para reducir dimensiones
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Visualizar los clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis', marker='o')
plt.title('K-Means Clustering with 3 Clusters (PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Crear un DataFrame con los loadings de cada variable para las dos primeras componentes
loadings = pd.DataFrame(pca.components_.T, index=X.columns, columns=['PC1', 'PC2'])
print(loadings)
```

```python
# Variables dominantes por componente
dominantes_PC1 = loadings['PC1'].abs().sort_values(ascending=False)
dominantes_PC2 = loadings['PC2'].abs().sort_values(ascending=False)

print("Variables más importantes en PC1:")
print(dominantes_PC1.head(5))  # top 5

print("\nVariables más importantes en PC2:")
print(dominantes_PC2.head(5))
```

**2. Diferencias observadas en el agrupamiento no supervisado**

Con KMeans y DBSCAN:

KMeans:

Crea clusters densos con un número fijo de grupos.

Agrupa a estudiantes según combinaciones de variables, sin usar la nota final.

Ejemplo: cluster de estudiantes con alta socialización y consumo de alcohol vs. estudiantes con hábitos de estudio más constantes.

DBSCAN:

**3. Coincidencias entre estudiantes que reprueban y clusters de bajo rendimiento**

Al comparar failures y notas bajas con los clusters de DBSCAN/KMeans:

Los outliers de DBSCAN a veces coinciden con estudiantes con combinaciones extremas de hábitos negativos (muchas reprobaciones, alto alcohol, poca asistencia).

**4. Uso combinado de ambos enfoques**

Supervisado + No supervisado puede ser muy útil para programas educativos:

Modelos supervisados:

Detectan riesgo de bajo rendimiento de forma individual.

Permiten priorizar estudiantes para tutorías o intervenciones tempranas.

Modelos no supervisados:

Detectan grupos de estudiantes con características similares.

Permiten diseñar programas de apoyo grupales, actividades sociales o talleres adaptados a cada cluster.

Los outliers de DBSCAN pueden recibir atención personalizada.

Ejemplo de aplicación:

Cluster de estudiantes con alto alcohol y bajas horas de estudio → taller de hábitos de estudio y concienciación sobre consumo.

Cluster de estudiantes con buen desempeño pero poco apoyo familiar → tutorías y mentorías para reforzar motivación.
